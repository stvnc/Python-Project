## Matplotlib Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
# from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [1, 3, 5, 2, 4, 6, 8, 7, 3]

plt.plot(x,y)
plt.title('X vs Y')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.grid(True)
plt.legend(['Garis XY'])
plt.show()
```

```python
'''
    ^ = Segitiga,
    s = Square,
    * = Bintang,
    o = Lingkaran besar,
    . = dot,
    .- =  dot garis,
'''
```

```python
plt.plot(x, x, color='r', marker='h', linestyle=':', markersize = 20)

plt.title('Testing')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y/Z')
plt.grid(True)
# plt.legend(['Garis XY', 'Garis XZ', 'Garis YZ'])

plt.savefig('figure.png')
plt.show()
```

```python
plt.plot(x, x, color='r', marker='h', markerfacecolor= 'k', markeredgecolor='b', linestyle=':',  
         markersize = 20, linewidth = .5, label = 'WoDeData')

plt.title('Testing')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y/Z')
plt.grid(True, color='salmon')
# plt.legend(['Garis XY', 'Garis XZ', 'Garis YZ'])

plt.savefig('figure.png')
plt.legend()
plt.show()
```

```python
x = np.array(x)
y = np.array(y)
plt.plot(x, x**2, color='r', marker='.', markerfacecolor= 'k', markeredgecolor='b', linestyle=':',  
         markersize = 6, linewidth = .5, label = 'WoDeData')
plt.plot(x, x**3, color='b', marker='.', markerfacecolor= 'k', markeredgecolor='r', linestyle=':',  
         markersize = 6, linewidth = .5, label = 'WoDeData2')

plt.title('Testing')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y/Z')
plt.grid(True, color='salmon')

for i in x**2:
    plt.text(x[list(x**2).index(i)]-.2, i+2, i)
for i in x**3: 
    plt.text(x[list(x**3).index(i)]-.2, i+2, i)
# plt.legend(['Garis XY', 'Garis XZ', 'Garis YZ'])

plt.savefig('figure.png')
plt.legend()
plt.show()
```
