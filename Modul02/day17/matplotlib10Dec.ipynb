## Matplotlib Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
# from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [1, 3, 5, 2, 4, 6, 8, 7, 3]

plt.plot(x,y)
plt.title('X vs Y')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.grid(True)
plt.legend(['Garis XY'])
plt.show()
```

```python
'''
    ^ = Segitiga,
    s = Square,
    * = Bintang,
    o = Lingkaran besar,
    . = dot,
    .- =  dot garis,
'''
```

```python
plt.plot(x, x, color='r', marker='h', linestyle=':', markersize = 20)

plt.title('Testing')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y/Z')
plt.grid(True)
# plt.legend(['Garis XY', 'Garis XZ', 'Garis YZ'])

plt.savefig('figure.png')
plt.show()
```

```python
plt.plot(x, x, color='r', marker='h', markerfacecolor= 'k', markeredgecolor='b', linestyle=':',  
         markersize = 20, linewidth = .5, label = 'WoDeData')

plt.title('Testing')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y/Z')
plt.grid(True, color='salmon')
# plt.legend(['Garis XY', 'Garis XZ', 'Garis YZ'])

plt.savefig('figure.png')
plt.legend()
plt.show()
```

```python
x = np.array(x)
y = np.array(y)
plt.plot(x, x**2, color='r', marker='.', markerfacecolor= 'k', markeredgecolor='b', linestyle=':',  
         markersize = 6, linewidth = .5, label = 'WoDeData')
plt.plot(x, x**3, color='b', marker='.', markerfacecolor= 'k', markeredgecolor='r', linestyle=':',  
         markersize = 6, linewidth = .5, label = 'WoDeData2')

plt.title('Testing')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y/Z')
plt.grid(True, color='salmon')

for i in x**2:
    plt.text(x[list(x**2).index(i)]-.2, i+2, i)
for i in x**3: 
    plt.text(x[list(x**3).index(i)]-.2, i+2, i)
# plt.legend(['Garis XY', 'Garis XZ', 'Garis YZ'])
maximumValue = np.amax(y**3)
plt.annotate(
    'Highest Value!', xy = (9, 729), xytext = (5, 600), #Lokasi highest value di 9, 729 dengan posisi text di 5, 600
    arrowprops = {
        'shrink': 0.1,
        'facecolor': 'red'} 
)
plt.savefig('figure.png')
plt.legend()
plt.show()
```

```python
'''
    3x + y = 15,
    x + y = 7
'''
firstEq = np.array([
    [3, 1],
    [1, 1]
])
secondEq = np.array([15, 7])

hasil = np.linalg.solve(firstEq, secondEq)
hasil
```

```python
firstEquation = np.array([3]).reshape(1, 1)
firstEquation2 = np.array([15]).reshape(1, 1)
firstValue = np.linalg.solve(firstEquation, firstEquation2)
firstValue = int(firstValue[0])
firstValue
```

```python
secondEquation = np.array([1]).reshape(1,1)
secondEquation2 = np.array([15]).reshape(1,1)
secondValue = np.linalg.solve(secondEquation, secondEquation2)
secondValue = int(secondValue[0])
secondValue
```

```python
thirdEquation = np.array([1]).reshape(1,1)
thirdEquation2 = np.array([7]).reshape(1,1)
thirdValue = np.linalg.solve(thirdEquation, thirdEquation2)
thirdValue = int(thirdValue[0])
thirdValue
```

```python
fourthEquation = np.array([1]).reshape(1,1)
fourthEquation2 = np.array([7]).reshape(1,1)
fourthValue = np.linalg.solve(fourthEquation, fourthEquation2)
fourthValue = int(fourthValue[0])
fourthValue
```

```python
firstLoc = [firstValue, 0]
secondLoc = [0, secondValue]
thirdLoc = [thirdValue, 0]
fourthLoc = [0, fourthValue]
plt.plot(firstLoc, secondLoc, thirdLoc, fourthLoc)
plt.plot(hasil[0], hasil[1], marker='*', markersize = 10, markerfacecolor='brown', markeredgecolor='white')
plt.annotate('titik potong', xy = (hasil[0], hasil[1]), xytext = (hasil[0]+1, hasil[1]+1),
            arrowprops={
                'facecolor': 'salmon',
                'shrink': .1
            })
plt.show()
```
