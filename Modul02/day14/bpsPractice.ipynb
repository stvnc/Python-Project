```python
import xlrd
import numpy as np
import pandas as pd
```

```python
bpsDf = pd.read_excel('databps.xlsx', na_values = ['-'], header=3, skipfooter=1)
```

```python
bpsDf.rename(columns = {'Unnamed: 0': 'Provinsi'}, inplace=True)
bpsDf
```

```python
listTahun = [2003, 2008, 2011, 2014, 2018]
listTahunString = ['2003.1', '2008.1', '2011.1', '2014.1', '2018.1']
for i in range(len(listTahun)):
    totalValue = bpsDf[listTahun[i]] + bpsDf[listTahunString[i]]
    bpsDf['Total'] = totalValue
    print(f'Desa Terbanyak pada Tahun {listTahun[i]} :')
    print(bpsDf[totalValue == totalValue.max()][["Provinsi", listTahun[i], listTahunString[i], 'Total']])
    
```

```python
listTahun = [2003, 2008, 2011, 2014, 2018]
listTahunString = ['2003.1', '2008.1', '2011.1', '2014.1', '2018.1']
for i in range(len(listTahun)):
    totalValue = bpsDf[listTahun[i]] + bpsDf[listTahunString[i]]
    bpsDf['Total'] = totalValue
    print(f'Desa Tersedikit pada Tahun {listTahun[i]} :')
    print(bpsDf[totalValue == totalValue.min()][["Provinsi", listTahun[i], listTahunString[i], 'Total']])
```
