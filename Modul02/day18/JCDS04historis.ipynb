```python
import numpy as np
import pandas as pd
```

```python
xlDf = pd.read_csv('EXCL.JK.csv', parse_dates = ['Date'], index_col = 'Date')
telkomselDf = pd.read_csv('TLKM.JK.csv', parse_dates = ['Date'], index_col = 'Date')
smartfrenDf = pd.read_csv('FREN.JK.csv', parse_dates = ['Date'], index_col = 'Date')
indosatDf = pd.read_csv('ISAT.JK.csv', parse_dates = ['Date'], index_col = 'Date')
```

```python
xlDf = xlDf.sort_index()
xlDf = xlDf.resample('D').sum()
xlDf.iloc[:, :5] = xlDf.iloc[:, :5].replace(0, np.NaN)
xlDf = xlDf.fillna(method = 'ffill')

telkomselDf = telkomselDf.sort_index()
telkomselDf = telkomselDf.resample('D').sum()
telkomselDf.iloc[:, :5] = telkomselDf.iloc[:, :5].replace(0, np.NaN)
telkomselDf = telkomselDf.fillna(method = 'ffill')

smartfrenDf = smartfrenDf.sort_index()
smartfrenDf = smartfrenDf.resample('D').sum()
smartfrenDf.iloc[:, :5] = smartfrenDf.iloc[:, :5].replace(0, np.NaN)
smartfrenDf = smartfrenDf.fillna(method = 'ffill')

indosatDf = indosatDf.sort_index()
indosatDf = indosatDf.resample('D').sum()
indosatDf.iloc[:, :5] = indosatDf.iloc[:, :5].replace(0, np.NaN)
indosatDf = indosatDf.fillna(method = 'ffill')
```

```python
tempList = list(xlDf.index.values)
dateList = []
for i in range(len(tempList)):
    dateList.append(pd.to_datetime(tempList[i]).strftime('%Y-%m-%d'))
    
dateList
```

```python
import matplotlib.pyplot as plt
```

```python
indexList = xlDf.index.tolist()
plt.figure(figsize = (11, 6), dpi = 80, facecolor = 'w', edgecolor = 'w')
plt.plot(indexList, xlDf['Close'], indexList, smartfrenDf['Close'], indexList, indosatDf['Close'], indexList, telkomselDf['Close'])
plt.legend(['PT XL Axiata Tbk', 'PT Smartfren Telecom Tbk', 'PT Indosat Tbk', 'PT Telekomunikasi Indonesia Tbk'], bbox_to_anchor=(0.,1,1., .102),
          loc = 'lower left', ncol = 4, mode = 'expand', borderaxespad = 0.)
plt.xticks(rotation = 45)
plt.xlabel('Tanggal', size = '17') 
plt.ylabel('Rupiah (IDR)', size = '17')
plt.grid(True)
plt.show()
```
