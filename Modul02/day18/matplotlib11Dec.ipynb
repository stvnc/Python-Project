```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'x' : [1, 2, 3, 4, 5],
    'y' : [9, 5, 7, 3, 6],
    'z' : [2, 3, 4, 5, 3]
})

plt.style.use('seaborn-dark')
plt.plot(df['x'], df[['y', 'z']])
plt.grid(True)
plt.title('Test Chart')
plt.xticks(np.arange(0, 5, step = 1), list('abcdef'), rotation = 0)
plt.yticks(np.arange(0, 10, step = 2), rotation = 0)
plt.show()
```

```python
plt.figure('Tes')

#Plot 1 di fig 2x2 = kiri atas
plt.subplot(221)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Test 1')

plt.subplot(224)
plt.plot([1, 2, 3], [3, 2, 1])
plt.title('Test 2')

plt.suptitle('Ini super title')

plt.show()
```

```python
plt.figure(facecolor = 'grey', edgecolor = 'black')
plt.bar(df.x, df.y, color = 'black', edgecolor = 'gold', linewidth = 2)
plt.title('Ini bar chart', fontdict = {
    'family':'impact', 'size':30,
    'color' : 'gold'
})
plt.xlabel('Ini X', color = 'gold')
plt.ylabel('Ini Y', color = 'gold')
plt.show()
```

```python
plt.bar(df.x, df.y, color = 'black',  edgecolor = 'gold', linewidth = 2)
# plt.bar(df.x, df.z, color = 'black', edgecolor = 'gold', linewidth = 2)

for i in df['y']:
    plt.text(
    df[df['y'] == i]['x'] -0.1, i+0, i,
    fontdict = {'size' : 30,
               'color':'gold'})
plt.yticks(np.arange(0,12,1))
plt.show()
```

```python
plt.bar(df.x, df.y, color = 'black', bottom = df.z, edgecolor = 'gold', linewidth = 2)
plt.bar(df.x, df.z, color = 'black', edgecolor = 'gold', linewidth = 2)

for i in df['y']:
    plt.text(
    df[df['y'] == i]['x'] -0.1, i, i,
    fontdict = {'size' : 30,
               'color':'gold'})
plt.yticks(np.arange(0,12,1))
plt.show()
```

```python
# plt.bar(df.x, df.y, color = 'black', bottom = df.z, edgecolor = 'gold', linewidth = 2)
plt.bar(df.z, df.x, color = 'black', edgecolor = 'gold', linewidth = 2)

for i in df['x']:
    plt.text(
    df[df['x'] == i]['z'] -0.1, i, i,
    fontdict = {'size' : 30,
               'color':'gold'})
plt.yticks(np.arange(0,12,1))
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 5, 6], 'gold', markersize = 20, zorder = 0)

plt.scatter(
    [1, 2, 3], [4, 5, 6], marker = '$§$', s = 300, color = 'k', edgecolor = 'gold', zorder = 3)
plt.bar([1, 2, 3], [4, 5, 6], zorder = 1, facecolor = 'k', edgecolor = 'gold', linewidth = 2)

plt.show()
```

```python
plt.figure(1).add_subplot(121)
plt.plot([1, 2, 3], [4, 5, 6], color = 'gold')
plt.title('test 1')
plt.figure(1).add_subplot(122)
plt.plot([1, 2, 3], [3, 2, 1], color = 'gold')
plt.title('test 2')
plt.suptitle('HELLO', color = 'gold')
plt.show()
```

```python
persentase = [5000, 7500, 4000, 7000, 10000]
keterangan = ['Pak A', 'Pak B', 'Pak C', 'Pak D', 'Pak E']

x,y,z = plt.pie(
    persentase, 
    labels = keterangan,
    colors = ['brown', 'gold', 'orange', 'salmon', 'purple'],
    startangle = 90, 
    shadow = True, 
    counterclock = False,
    textprops = { 'size' : 9, 'color' : '#3e2f2e'},
    autopct = '%1.1f%%'
)

for i in z:
    i.set_color('w')
plt.show()
```
