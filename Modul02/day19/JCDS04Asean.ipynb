```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlalchemy as sqlalc
```

```python
alamatDb = "mysql://root:vincent28@localhost:3306/world"
engine = sqlalc.create_engine(alamatDb)
query = 'select Name as nama_negara, GNP, Population, SurfaceArea as luas_daratan  from country where region = "Southeast Asia" order by nama_negara'
dfAlc = pd.read_sql(query, engine)
dfAlc
```

```python
plt.figure(figsize = (14, 6), dpi = 80)
plt.title('Populasi Negara ASEAN', fontsize=20)
plt.bar(dfAlc.nama_negara, dfAlc.Population, linewidth = 3,
       color = ['brown', 'gold', 'orange', 'salmon', 'purple', 'red', 'blue', 'black', 'grey', 'green', 'cyan'])
for i in dfAlc['Population']:
    plt.text(
    dfAlc[dfAlc['Population'] == i]['nama_negara'], i, i, ha ='center', va = 'bottom')
plt.xticks(dfAlc.nama_negara, rotation = 45)
plt.xlabel('Negara', size = '17') 
plt.ylabel('Populasi (x100jt jiwa)', size = '17')
plt.grid(True, color = 'grey')
```

```python
x,y,z = plt.pie(
    dfAlc.Population, 
    labels = dfAlc.nama_negara,
    colors = ['brown', 'gold', 'orange', 'salmon', 'purple', 'red', 'navy', 'black', 'grey', 'green', 'cyan'],
    startangle = 0, 
    shadow = True, 
    radius = 2.5,
    counterclock = True,
    textprops = { 'size' : 12, 'color' : '#3e2f2e'},
    autopct = '%1.1f%%'
)

for i in z:
    i.set_color('w')
```

```python
plt.figure(figsize = (14, 6), dpi = 80)
plt.title('Pendapatan Bruto Nasional ASEAN', fontsize=20)
plt.bar(dfAlc.nama_negara, dfAlc.GNP, linewidth = 3,
       color = ['brown', 'gold', 'orange', 'salmon', 'purple', 'red', 'navy', 'black', 'grey', 'green', 'cyan'])
for i in dfAlc['GNP']:
    plt.text(
    dfAlc[dfAlc['GNP'] == i]['nama_negara'], i, i, ha = 'center', va = 'bottom')
plt.xticks(dfAlc.nama_negara, rotation = 45)
plt.xlabel('Negara', size = '17') 
plt.ylabel('Gross National Product(US$)', size = '17')
plt.grid(True, color = 'grey')
```

```python
x,y,z = plt.pie(
    dfAlc.luas_daratan, 
    labels = dfAlc.nama_negara,
    colors = ['brown', 'gold', 'orange', 'salmon', 'purple', 'red', 'navy', 'black', 'grey', 'green', 'cyan'],
    startangle = 0,
    radius = 2.5,
    shadow = True, 
    counterclock = True,
    textprops = { 'size' : 9, 'color' : '#3e2f2e'},
    autopct = '%1.1f%%'
)

for i in z:
    i.set_color('w')
```
