```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as im
```

```python
dfSales = pd.read_csv('dataSales.csv')
dfSales
```

### Pivot

```python
dfSales = dfSales.pivot(index = 'nama', columns = 'mobil')
dfSales
```

```python
fig, p = plt.subplots()
plt.imshow(dfSales, cmap = 'BuPu_r')
plt.colorbar()

col = list(map(lambda x: x[1], dfSales.columns.tolist()))
col
col2 = list(map(lambda x: x, dfSales.index.tolist()))
col2

for i in range(len(col)):
    for j in range(len(col2)):
        plt.text(j, i, dfSales.iloc[i, j],
                color = 'black', fontsize = 20,
                ha = 'center', va = 'center')

p.set_xticks(np.arange(len(col)))
p.set_xticklabels(col)
p.set_yticks(np.arange(len(col2)))
p.set_yticklabels(col2)
p.set_ylim(2.5,-0.5)

plt.show()
```

### Atlit.csv

```python
dfAtlit = pd.read_csv('atlit.csv')
dfAtlit.corr()
```

```python
plt.imshow(dfAtlit.corr(), cmap = 'magma_r')
plt.colorbar()
col = list(map(lambda x: x, dfAtlit.corr().columns.tolist()))

plt.xticks(
    np.arange(len(dfAtlit.corr().index)),
    dfAtlit.corr().index.tolist(), rotation = 45)

plt.yticks(
    np.arange(len(dfAtlit.corr().index)),
    dfAtlit.corr().index.tolist())

for x in range(len(col)):
    for y in range(len(col)):
        plt.text(y, x, round(dfAtlit.corr().iloc[x, y], 3),
                color = 'orange',
                ha = 'center', va = 'center')
        
plt.ylim(-.5, 4.5)
plt.show()
```
