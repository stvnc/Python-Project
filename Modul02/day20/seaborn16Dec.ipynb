```python
df = sb.load_dataset('flights')
df = df.pivot('month', 'year', 'passengers') #month menjadi index, year menjadi kolom, dan passengers menjadi valuenya
df = df.sum()
df = df.to_frame()
df
```

```python
df.values
```

```python
sb.barplot(data = df, x = df.index, y = 0)
```

```python
df2 = sb.load_dataset('flights')
df2 = df2.pivot('month', 'year', 'passengers')
sb.heatmap(df2, annot=True, cmap='magma_r')
```

```python
fig = plt.figure()
p = plt.subplot(111, projection= "3d")

x = np.arange(10)
y = np.arange(10)
z = np.zeros(10)

dx = np.ones(10)
dy = np.ones(10)
dz = np.arange(10)

p.bar3d(x, y, z, dx, dy, dz)

p.set_xlabel("Sumbu X")
p.set_ylabel("Sumbu Y")
p.set_zlabel("Sumbu Z")

plt.show()
```
