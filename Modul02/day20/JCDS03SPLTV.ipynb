```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

### Soal 2 - SPLTV (Sistem Persamaan Linear Tiga Variabel)

Terdapat 3 buah persamaan linear 3 variabel sebagai berikut:

x  - 2y +  z =  6
3x +  y - 2z =  4
7x - 6y -  z = 10
Dengan menggunakan numpy, buatlah sebuah program python yang dapat menentukan nilai x, y & z dari 3 buah persamaan di atas! Hasil yang diharapkan saat program dieksekusi adalah:

Nilai x = 5.0
Nilai y = 3.0
Nilai z = 7.0



```python
findDotEq = np.array([
    [1, -2, 1],
    [3, 1, -2],
    [7, -6, -1]
    ])

valueEq = np.array([6, 4, 10])

hasil = np.linalg.solve(findDotEq, valueEq)
hasil
```

## 1st plot

```python
firstXEq = np.array([1]).reshape(1,1)
firstYEq = np.array([-2]).reshape(1,1)
firstZEq = np.array([1]).reshape(1,1)
firstVal = np.array([6]).reshape(1,1)
```

```python
firstX = np.linalg.solve(firstXEq, firstVal)
firstX = np.array([firstX, 0, 0])
firstY = np.linalg.solve(firstYEq, firstVal)
firstY = np.array([0, firstY, 0])
firstZ = np.linalg.solve(firstZEq, firstVal)
firstZ = np.array([0, 0 , firstZ])
```

```python
fig = plt.figure(figsize = (15, 9))
ax = fig.add_subplot(111, projection='3d')
firstPoint = [list(zip(firstX, firstY, firstZ))]
firstPolygon = Poly3DCollection(firstPoint, alpha=.55, facecolor='r')
plt.gca().add_collection3d(firstPolygon)
ax.plot3D(firstX, firstY, firstZ, 'red')
ax.set_xlabel('Nilai X')
ax.set_ylabel('Nilai Y')
ax.set_zlabel('Nilai Z')
plt.title('x-2y+z = 6', size = 20)
plt.show()
```
