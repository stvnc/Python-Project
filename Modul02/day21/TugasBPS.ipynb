```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
bpsDf = pd.read_excel('penduduk.xls', index_col = 'Unnamed: 0', na_values = '-', header = 3, skipfooter = 3)
bpsDf.fillna(0, inplace=True)
bpsDf
```

```python
fig = plt.figure(figsize = (20, 15))
p = plt.subplot(111, projection= "3d")

listTahun = [1971, 1980, 1990, 1995, 2000, 2010]
        
x = np.arange(len(bpsDf.index))
z = np.zeros(len(bpsDf.index))

dx = np.ones(len(bpsDf.index))
dy = np.ones(len(bpsDf.index))

for i in range(len(listTahun)):
    y = []
    dz = []
    y = [i-0.2] * len(bpsDf.index)
    for j in range(len(bpsDf.index)):
        dz.append(bpsDf.iloc[j, i])
    p.bar3d(x, y, z, dx, dy, dz)

plt.title('Data BPS', size = 27)
p.set_xlabel("Provinsi", size = 15)
p.set_ylabel("Tahun", size = 15)
p.set_zlabel("Populasi", size = 15)
p.set_xticklabels(bpsDf.index, rotation = 90)
p.set_xticks(range(len(bpsDf.index)))
p.set_yticklabels(listTahun, rotation = 90)
p.set_yticks(range(len(listTahun)))
plt.grid(True, color = 'salmon')
plt.show()
```
