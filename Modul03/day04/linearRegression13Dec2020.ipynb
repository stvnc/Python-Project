### Linear Regression

Gradient/slope $\displaystyle m = \frac {n(\sum xy) - (\sum x)(\sum y)}{n(\sum x^2) - (\sum x)^2}$

```python
### Multivariate Linear Regression

Multiple variable linear regression

$$\hat{y} = m_0 x_0 + m_1 x_1 + m_2 x_2 + ... + c$$
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'mesin x1' : [112, 250, 360, 450, 600],
    'usia x2' : [5, 3, 3, 1, 1],
    'km x3': [8, 7, 7, 5, 4],
    'harga y' : [100, 200, 300, 400, 500]
}
df = pd.DataFrame(data)
df.corr()
```

```python
model = LinearRegression()
model.fit(df[['mesin x1', 'usia x2', 'km x3']], df['harga y'])
```

```python
m = model.coef_ #Urutan Mesin, Usia, KM
c = model.intercept_
```

```python
plt.figure(figsize = (16, 8))
plt.subplot(131)
plt.plot(df['mesin x1'], df['harga y'], 'ro-')
plt.subplot(132)
plt.plot(df['usia x2'], df['harga y'], 'bo-')
plt.subplot(133)
plt.plot(df['km x3'], df['harga y'], 'go-')
plt.show()
```

```python
#Mesin, usia, km
model.predict([[112, 5, 8]])
```

```python
df['y"x1'] = m[0] * df['mesin x1'] + c
df['y"x2'] = m[1] * df['usia x2'] + c
df['y"x3'] = m[2] * df['km x3'] + c
df
```

```python
plt.figure(figsize = (16, 8))
plt.subplot(131)
plt.plot(df['mesin x1'], df['harga y'], 'ro-')
plt.plot(df['mesin x1'], df['y"x1'], 'yo-')
plt.subplot(132)
plt.plot(df['usia x2'], df['harga y'], 'bo-')
plt.plot(df['usia x2'], df['y"x2'], 'yo-')
plt.subplot(133)
plt.plot(df['km x3'], df['harga y'], 'go-')
plt.plot(df['km x3'], df['y"x3'], 'yo-')
plt.show()
```
