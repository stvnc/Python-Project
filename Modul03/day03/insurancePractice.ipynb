```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lr
```

```python
insuranceDf = pd.read_csv('insurance.csv')
insuranceDf.rename(columns = {'sex':'gender'}, inplace = True)
insuranceDf
```

```python
genderDict = {
    'female': 0,
    'male' : 1
}

smokerDict = {
    'no' : 0,
    'yes' : 1
}

for i in range(len(insuranceDf.index)):
    insuranceDf.iloc[i, 1] = genderDict.get(insuranceDf.iloc[i, 1])
    insuranceDf.iloc[i, 4] = smokerDict.get(insuranceDf.iloc[i, 4])
insuranceDf
```

```python
model = lr()
model.fit(insuranceDf[['age', 'gender', 'bmi', 'smoker']], insuranceDf['charges'])

print(model.coef_)
print(model.intercept_)
```

```python
print(f'prediksi data untuk wanita berumur 20 tahun, dengan BMI 20 dan bukan perokok: {model.predict(np.array([[20, 0, 20, 0]]))[0]}')
```
