```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
data = {
    'mesin': [1000, 2000, 3000, 4000, 5000],
    'harga': [10, 25, 35, 55, 80]
}
df = pd.DataFrame(data)
df
```

```python
tempData = []
for i in df['mesin']:
    tempData.append(i*i)
df['x^2'] = tempData
```

```python
tempData = []
for i in df['harga']:
    tempData.append(i*i)
df['y^2'] = tempData
```

```python
tempData = []
for i in range(len(df['mesin'])):
    tempData.append(df['mesin'].iloc[i]*df['harga'].iloc[i])
df['xy'] = tempData
```

```python
df
```

```python
#Slope = =((5*E7)-(A7*B7))/((5*C7)-(A7^2))
slope = ((len(df['mesin'])*(df['xy'].sum()))-(df['mesin'].sum()*(df['harga'].sum())))/((len(df['mesin'])*(df['x^2'].sum()))-(np.square(df['mesin'].sum())))
slope
```

```python
#intercept =((B7*C7)-(A7*E7))/((COUNT(A2:A6)*C7)-(A7)^2)
intercept = (((df['harga'].sum())*(df['x^2'].sum()))-(df['mesin'].sum()*(df['xy'].sum())))/((len(df['mesin'])*(df['x^2'].sum())) - np.square(df['mesin'].sum()))
intercept
```

```python
yAks = []

for i in df['mesin']:
    yAks.append(slope * i + intercept)
yAks
```

```python
df["y'"] = yAks
df
```

```python
yMinYaks = []

for i in range(len(df['harga'])):
    yMinYaks.append(np.abs(df['harga'].iloc[i] - df["y'"].iloc[i]))
yMinYaks
```

```python
df["|y-y'|"] = yMinYaks
df
```

```python
sqrYMinYaks = []
for i in df["|y-y'|"]:
    sqrYMinYaks.append(np.square(i))
sqrYMinYaks
```

```python
df["|y-y'|^2"] = sqrYMinYaks
df
```

```python
selisih = []
for i in range(len(df['harga'])):
    selisih.append(np.abs(df['harga'].iloc[i] - np.mean(df['harga'])))
selisih
```

```python
df['selisih |y-avg(y)|'] = selisih
df
```

```python
sqrSelisih = []
for i in df['selisih |y-avg(y)|']:
    sqrSelisih.append(np.square(i))
sqrSelisih
```

```python
df['selisih^2'] = sqrSelisih
df
```

```python
eLog = []
for i in df['harga']:
    eLog.append(np.log(1 + i))
eLog
```

```python
df['elog(1+y)'] = eLog
df
```

```python
eLogAcc = []
for i in df["y'"]:
    eLogAcc.append(np.log(1 + i))
eLogAcc
```

```python
df["elog(1+y')"] = eLogAcc
df
```

```python
eLogVal = []
for i in range(len(df['elog(1+y)'])):
    eLogVal.append(np.square(df['elog(1+y)'].iloc[i] - df["elog(1+y')"].iloc[i]))
eLogVal
```

```python
df["(elog(1+y) - elog(1+y'))^2"] = eLogVal
df
```

```python
percentageOfError = []
for i in range(len(df["|y-y'|"])):
    percentageOfError.append(np.abs(df["|y-y'|"].iloc[i]/df['harga'].iloc[i]))
percentageOfError
```

```python
df['%error'] = percentageOfError
df
```

```python
maxOfError = np.max(df["|y-y'|"])
meanAbsoluteError = (df["|y-y'|"].sum())/len(df["mesin"])
meanSquaredError = np.square(meanAbsoluteError)
squareRootMSE = np.sqrt(meanSquaredError)
medianAbsoluteError = np.median(df["|y-y'|"])
meanSquaredLogarithmicError = (df["(elog(1+y) - elog(1+y'))^2"].sum())/len(df["(elog(1+y) - elog(1+y'))^2"])
squaredMSLE = np.sqrt(meanSquaredLogarithmicError)
r2Score = 1-(df["|y-y'|^2"].sum())/(df['selisih^2'].sum())
meanAbsolutePercentageError = df['%error'].sum()/len(df['%error'])
print("Max Of Error: ", round(maxOfError, 2))
print("Mean Absolute Error: ", round(meanAbsoluteError,2))
print("Mean Squared Error: ", round(meanSquaredError,2))
print("Root Squared Mean Squared Error: ", round(meanSquaredError,2))
print("Median Absolute Error: ", round(medianAbsoluteError,2))
print("Mean Squared Logarithmic Error: ", round(meanSquaredLogarithmicError, 2))
print("Square MSLE: ", round(squaredMSLE, 2))
print("R2 Score: ", round(r2Score, 2))
print("Mean Absolute Percentage Error: ", round(meanAbsolutePercentageError, 2))
```
